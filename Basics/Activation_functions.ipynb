# Activation Functions




1. **Sigmoid Function**:
   $
   \sigma(x) = \frac{1}{1 + e^{-x}}
   $

2. **Hyperbolic Tangent Function (tanh)**:
   $
   \tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}
   $

3. **Rectified Linear Unit (ReLU)**:
   $
   \text{ReLU}(x) = \max(0, x)
   $

4. **Leaky Rectified Linear Unit (Leaky ReLU)**:
   $
   \text{Leaky ReLU}(x) = \begin{cases}
      \alpha x & \text{if } x < 0 \\
      x & \text{if } x \geq 0
   \end{cases}
   $
   
   where $\alpha$ is a small positive constant (e.g., 0.01).

5. **Softmax Function**:
   $
   \text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}
   $


   where $x_i$ is the $i$-th element of the input vector $x$, and $n$ is the total number of elements in $x$.

# Parametrized sigmoid
$$
S_{(b,a,t,p)}(x) =p+ \frac{b}{1+e^{-a(x-t)}} \\
\text{default values: } a=1, b=1, t=0, p=0
$$

$$
y_{tower} =
S_{(b=100,a=300,t=0,p=0)}(x)
+ S_{(b=100,a=-300,t=5,p=-B)}(x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x,a=1,b=1,t=0,p=0):
    return p + b / (1 + np.exp(-a*(x-t)))

# Values
x = np.linspace(-10, 10, 1000)
A=300
B=100
y1 = sigmoid(x,a=A,b=B,t=0,p=0)
y2 = sigmoid(x,a=-A,b=B,t=2,p=-B)*0
y=y1+y2

# Plot
plt.plot(x, y, c='red', linewidth=5)
plt.title('Combination of Sigmoid Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

# The hyperbolic tangent function

$$\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $$



```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbolic tangent function
def tanh(x):
    return np.tanh(x)

# Values
x = np.linspace(-10, 10, 1000)
y = tanh(x)

# Plot
plt.plot(x, y)
plt.title('Hyperbolic Tangent Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
