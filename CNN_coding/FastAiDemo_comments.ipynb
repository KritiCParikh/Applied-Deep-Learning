```python
path = untar_data(URLs.CIFAR)

#Unzip the dataset.
```

```python
data = DataBlock(blocks=(ImageBlock(), CategoryBlock()),
                    get_items=get_image_files,
                    get_y=parent_label,
                    item_tfms=Resize(40))
```

```python
#Created an image folder and specifying the validation split between the train and the validation. bs= batch size
dls = data.dataloaders(path, bs=64, valid_pct=0.2, seed=42,)
#Dataloader= a class with our images and their labels.
```

```python
dls.valid.show_batch(max_n=12, nrows=3)
```

```python
learner = vision_learner(dls, vgg16_bn, metrics=accuracy)
learner.fine_tune(10)
#Fine tuning for 10 epochs.
```
